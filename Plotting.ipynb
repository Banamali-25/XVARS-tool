# R2 Plots #Manuscript Figure no. 5

```python
# To obtain the visuals of training and testing scores for 30 runs

import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot
import seaborn
from matplotlib.ticker import FormatStrFormatter
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
Data = pd.read_csv('path\R2_Plot.csv')
dataset = pd.DataFrame(Data)
dataset
```

# R2 Plots for NH4+ and Chl-a predictions

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5),sharey=True)
plt.subplots_adjust(wspace=0.01, hspace=0.2)

ax1 = sns.swarmplot(x="Models", y="R2_score", hue="Type",data=dataset,ax=ax1, palette=["violet", "k"], color=".15")

ax1.set_ylabel('$\mathbf{R^2}$', fontsize=20,weight='bold')
ax1.set_yticklabels(ax1.get_yticks(), fontsize = 11, fontweight='bold',color='black')
ax1.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
ax1.legend_.remove()
ax1.set_title(r'$\mathbf{NH}_4^+$', fontsize=20,fontweight = 'bold')
ax1.xaxis.set_tick_params(labelsize = 12, labelcolor ='black')
ax1.set(xlabel=None)

ax2 = sns.swarmplot(x="Models.1", y="R2_score.1", hue="Type.1",data=dataset,ax=ax2, palette=["violet", "k"], color=".15")
ax2.set(xlabel=None)
ax2.set(ylabel=None)
ax2.xaxis.set_tick_params(labelsize = 12, labelcolor ='black')
ax2.legend_.remove()
ax2.set_title('Chl-a', fontsize=20,fontweight = 'bold')
```

# R2 Plots for DO and pH predictions

```python
#R2 plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5),sharey=True)
plt.subplots_adjust(wspace=0.01, hspace=0.2)


ax1 = sns.swarmplot(x="Models.2", y="R2_score.2", hue="Type.2",data=dataset,ax=ax1, palette=["violet", "k"], color=".25")
ax1.set_xlabel('Models', fontsize=18,weight='bold');
ax1.set_ylabel('$\mathbf{R^2}$', fontsize=20,weight='bold');
ax1.set_yticklabels(ax1.get_yticks(), fontsize = 11, fontweight='bold',color='black')
ax1.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
ax1.legend_.remove()
ax1.set_title('DO', fontsize=20,fontweight = 'bold')
ax1.xaxis.set_tick_params(labelsize = 12, labelcolor ='black')

ax2 = sns.swarmplot(x="Models.3", y="R2_score.3", hue="Type.3",data=dataset,ax=ax2,palette=["violet", "k"],color=".25")
ax2.set_xlabel('Models', fontsize=18,weight='bold');
ax2.xaxis.set_tick_params(labelsize = 12, labelcolor ='black')
ax2.set(ylabel=None)
ax2.set_title('pH', fontsize=20,fontweight = 'bold')
handles, labels = ax2.get_legend_handles_labels()
plt.legend( numpoints=1, labelspacing=0.5,bbox_to_anchor = (0.01, 0.01),loc='lower left', fontsize=20)
```
